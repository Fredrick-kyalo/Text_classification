```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate some sample data for the linear regression model
np.random.seed(38)  # Set a seed for reproducibility
x = np.random.rand(200)
print(x)
y=0.5*x *np.random.random_sample(200)
print(y)
```

```python
#Gradient Descent
slope = 0
intercept = 0
n=len(x)


learn_rate = 0.001;
itterations = 900
```

```python
for i in range(itterations):
  y_predicted = slope * x + intercept

  slopederivative = -(2/n)*sum(x*(y-y_predicted))
  xderivative = -(2/n)*sum(y-y_predicted)
  cost = np.mean((y_predicted - y)**2)
  slope = slope - learn_rate* slopederivative
  intercept= intercept - learn_rate * xderivative

```

```python
for i in range(len(x)):
  print("Actual(y): " ,y[i],"predicted (y_pred):" ,y_predicted[i],"Diff:",y[i]-y_predicted[i])
```

```python
print(" ïtterations ",i+1,"|MSE: ",(cost,2),"| slope:",(slope,2),"|Intercept:",(intercept,2))
print("---------------------------------------" )
plt.scatter(x,y)
plt.plot(x,y_predicted,color="green")
plt.show()
```
